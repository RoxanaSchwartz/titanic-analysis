```python
# Load libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("Libraries loaded!")
```

```python
# load data
df = pd.read_csv("train.csv")
print(df.head())
```

```python
# Survival reate by gender
sns.barplot(x = "Sex", y = "Survived", data = df)
plt.title("Survival rate by gender")
plt.show()
```

```python
df.to_excel("titanic_output.xlsx", index = False)
print("Data exported to Excel!")
```

```python
#Survival rate by class
sns.barplot(x="Pclass", y="Survived", data=df)
plt.title("Survival rate by passenger class")
plt.show()
```

```python
# Histogram: Age of survivors vs. non-survivors
plt.hist(df[df["Survived"] == 1]["Age"], bins=20, alpha=0.5, label="survived")
plt.hist(df[df["Survived"] == 0]["Age"], bins=20, alpha=0.5, label="not survived")
plt.title("Distribution of age by survival")
plt.xlabel("Age")
plt.ylabel("Number")
plt.legend()
plt.show()
```

```python
# Survival rate by gender and class
sns.catplot(x="Pclass", y="Survived", hue="Sex", kind="bar", data=df)
plt.title("Survival rate by class and gender")
plt.show()
```

```python
# Show missing values
print("Missing values per column:")
print(df.isnull().sum())
```

```python
# Fill missing age values with the median
df["Age"] = df["Age"].fillna(df["Age"].median())
print("Missing age values filled. New missing values:")
print(df.isnull().sum())
```

```python
# Analyze survival rate by embarkation location
sns.barplot(x="Embarked", y="Survived", data=df)
plt.title("Survival rate by embarkation location")
plt.xlabel("Embarkation location (S = Southampton, C = Cherbourg, Q = Queenstown)")
plt.show()
```

```python
# Analyze fare distribution by survival
sns.boxplot(x="Survived", y="Fare", data=df)
plt.title("Fare distribution based on survival")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Fare")
plt.show()
```

```python
plt.hist(df[df["Survived"] == 1]["Fare"], bins=20, alpha=0.5, label="Survived")
plt.hist(df[df["Survived"] == 0]["Fare"], bins=20, alpha=0.5, label="Not survived")
plt.title("Fare distribution based on survival")
plt.xlabel("Fare")
plt.ylabel("Number")
plt.legend()
plt.show()
```

```python
# Export data to Excel
df.to_excel("titanic_output.xlsx", index=False)
print("Exported data to titanic_output.xlsx!")
```
